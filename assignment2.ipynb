```python
def signs_helper(x):
    try: 
        if "it doesn't matter" in x:
            return "doesn't matter"
        elif "it matters a lot" in x:
            return "matters"
        elif "fun" in x:
            return "fun to think about"
        else:
            return "N/A"
    except:
        return "N/A"
        
    
```

```python
df['signmod'] = df['sign'].apply(signs_helper)
```

```python
sns.set(rc={'figure.figsize':(20,10)})

plt = sns.countplot(x = "signmod", data = df, hue = 'sex', palette=["b", "g"])

plt.set_xticklabels(plt.get_xticklabels(),rotation=90)
```
